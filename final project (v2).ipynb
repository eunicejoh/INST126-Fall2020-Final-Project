```python
# import pandas -- Eunice 
import pandas as pd

# import time for welcome/goodbye messages -- Jalen
import time

# import matplotlib for visualizations -- Eunice 
import matplotlib.pyplot as plt

# read csv -- Eunice
cereals = pd.read_csv("cereal.csv")

# rename column names for consistency and clearer understanding -- Eunice
cereals.rename(columns = {'name':'Name of Cereal', 'calories':'Calories per serving', 
                          'protein': 'Protein(g)', 'fat': 'Fat(g)', 'sodium':'Sodium (mg)', 
                          'fiber':'Fiber(g)', 'carbo': 'Carbohydrates(g)', 'sugars':'Sugars(g)', 
                          'potass':'Potassium(mg)','vitamins': 'Vitamins %', 'rating': 'Rating'},
                          inplace = True)

# get rid of "..."s to display the entire dataset at once -- Zijing 
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('max_colwidth',100)

# delete unnecessary columns -- Eunice
del cereals['shelf']
del cereals['weight']
del cereals['cups']
del cereals['mfr']
del cereals['type']
del cereals['Unnamed: 0']
del cereals['Unnamed: 0.1']
```

```python
# question 2 function -- Zijing
def compare_cereal():
    # set default answer to "yes" so the loop works
    answer = "yes"
    
    while answer == "yes":
        # menu display
        display(cereals)
        print()# just an empty line
        
        # tell users to input two cereals they want to compare
        cereal_1 = str(input("Which two cereals do you want to compare?\n First cereal: "))
        cereal_2 = str(input("Second cereal: "))
        
        # make a list of all the cereal names
        cereal_list = cereals['Name of Cereal'].tolist()
        
        # input check
        while not (cereal_1 in cereal_list) or not (cereal_2 in cereal_list):
            print("Please check your input.\n")
            cereal_1 = str(input("Which two cereals do you want to compare?\n First cereal: "))
            cereal_2 = str(input("Second cereal: "))
        
        # print two products information for comparation
        # First draft (ignore)
        # display(cereals.query('`Name of Cereal`==@cereal_1'))
        # display(cereals.query('`Name of Cereal`==@cereal_2'))
        
        # display charts(visualization of data)  
        compare_data = cereals.query('`Name of Cereal`== @cereal_1 or `Name of Cereal`==@cereal_2')
        
        # display two products information for comparison
        display(compare_data)
        
        # display all the visual comparison of nutrient between two chosen cereals
        plt.bar(compare_data['Name of Cereal'],compare_data['Calories per serving'])
        plt.title("Calorie Comparation")
        plt.xlabel("Name of Cereal")
        plt.ylabel("Calories per serving")
        plt.show()
        
        plt.bar(compare_data['Name of Cereal'],compare_data['Protein(g)'])
        plt.title("Protein Comparation")
        plt.xlabel("Name of Cereal")
        plt.ylabel("Protein(g)")
        plt.show()
        
        plt.bar(compare_data['Name of Cereal'],compare_data['Fat(g)'])
        plt.title("Fat Comparation")
        plt.xlabel("Name of Cereal")
        plt.ylabel("Fat(g)")
        plt.show()
        
        plt.bar(compare_data['Name of Cereal'],compare_data['Sodium (mg)'])
        plt.title("Sodium Comparation")
        plt.xlabel("Name of Cereal")
        plt.ylabel("Sodium (mg)")
        plt.show()
        
        plt.bar(compare_data['Name of Cereal'],compare_data['Fiber(g)'])
        plt.title("Fiber Comparation")
        plt.xlabel("Name of Cereal")
        plt.ylabel("Fiber(g)")
        plt.show()
        
        plt.bar(compare_data['Name of Cereal'],compare_data['Carbohydrates(g)'])
        plt.title("Carbohydrates Comparation")
        plt.xlabel("Name of Cereal")
        plt.ylabel("Carbohydrates(g)")
        plt.show()
        
        plt.bar(compare_data['Name of Cereal'],compare_data['Sugars(g)'])
        plt.title("Sugars Comparation")
        plt.xlabel("Name of Cereal")
        plt.ylabel("Sugars(g)")
        plt.show()
        
        plt.bar(compare_data['Name of Cereal'],compare_data['Potassium(mg)'])
        plt.title("Potassium Comparation")
        plt.xlabel("Name of Cereal")
        plt.ylabel("Potassium(mg)")
        plt.show()
        
        plt.bar(compare_data['Name of Cereal'],compare_data['Vitamins %'])
        plt.title("Vitamins % Comparation")
        plt.xlabel("Name of Cereal")
        plt.ylabel("Vitamins %")
        plt.show()
        
        plt.bar(compare_data['Name of Cereal'],compare_data['Rating'])
        plt.title("Overall Rating Comparation")
        plt.xlabel("Name of Cereal")
        plt.ylabel("Rating")
        plt.show()
        
        # see if the user want to go over again
        answer = str(input("Do you want to see the menu again?\n Input 'Yes' for menu, any other key to continue\n: "))
```
